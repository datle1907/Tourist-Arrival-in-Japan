```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import lightgbm as lgm
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import datetime as dt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
%matplotlib inline
```

## This is a regression task to predict "Tourist Arrivals in Japan".

```python
# Read file
data_train = pd.read_csv('train_df.csv')
data_train
```

```python
# Show information about data for each columns
data_train.info()
```

```python
# Columns having missing value
data_train.isna().sum()
```

```python
# Conver date to datetime
data_train['date'] = pd.to_datetime(data_train['date'])
data_train
```

```python
# Create new features
data_train['day'] = data_train['date'].dt.day
data_train['month'] = data_train['date'].dt.month
data_train['year'] = data_train['date'].dt.year
```

```python
# Drop columns date
data_train.drop(['id','date'], axis = 1, inplace = True)
data_train
```

```python
# Plot boxplot
numerical_col = data_train.select_dtypes(include = ['int64', 'float64'])
fig, axes = plt.subplots(3,3, figsize = (20,15))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Iterate through numerical columns and create boxplots
for i,col in enumerate(numerical_col.columns):
    sns.boxplot(data = numerical_col[col], ax = axes[i], palette = 'husl')
    axes[i].set_title(f'{col}')
    axes[i].set_xlabel('Value')

# Remove empty subplots
if len(numerical_col.columns) % 3 != 0:
    for i in range(len(numerical_col.columns), 3*3):
        fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

```python
# Create def to remove outliers
def detect_outlier(series, threshold = 3):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - threshold * iqr
    upper_bound = q3 + threshold * iqr
    return (series < lower_bound) | (series > upper_bound)


# Define the threshold for identifying outliers
outlier_threshold = 3

# Define columns to check
column_to_check = ['tourist_arrivals','tourism_index']

# Before identifying outliers, print the total row count
total_rows_before = len(data_train)
print(f"Total rows before outlier identification: {total_rows_before}")

# Find rows with outliers in the specified columns and count them
outlier_rows_before = data_train[column_to_check].apply(detect_outlier, threshold = outlier_threshold, axis=1)
total_outliers_before = outlier_rows_before.any(axis=1).sum()
print(f"Total rows with outliers before removal: {total_outliers_before}")

# Remove row with outliers
data_train = data_train[~outlier_rows_before.any(axis=1)]

# After removing rows with outliers, print the total row count
total_rows_after = len(data_train)
print(f"Total rows after removing outliers: {total_rows_after}")
```

```python
# Select columns contains object
cat_columns = data_train.columns[data_train.dtypes == 'object'].values.tolist()
cat_columns
```

```python
# Label encoder categorical columns
for cate in cat_columns:
    label_train = LabelEncoder()
    label_train = label_train.fit(data_train[cate])
    data_train[cate] = label_train.transform(data_train[cate]) # Encoding columns
    data_train[cate] = data_train[cate].astype('category')
```

```python
# Define features and target
y = data_train['tourist_arrivals']
X = data_train.drop(['tourist_arrivals'], axis = 1)
```

```python
# Split data into train, validation, and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Hyperparameter Tuning

### Trường hợp không sử dụng phương pháp Hyperparameter tuning

Sau khi chia tập data thành các tập train, set, validation xong thì chúng ta sẽ thử không dùng thư viện Optuna để tìm ra best hyper-parameter. Để chúng ta tìm ra sự khác biệt giữa Tuning hyperparameters và không Tuning Hyperparameters. Đầu tiên chúng ta sẽ thử không dùng thư viện Optuna để Fine Tune.


```python
lgb_regressor = lgb.LGBMRegressor(
    objective='regression',
    boosting_type='gbdt',
    num_leaves=10,
    learning_rate=0.1,
    n_estimators=50,
    min_data_in_leaf = 10,
    feature_fraction=0.4,
    bagging_fraction=0.3,
    bagging_freq=1,
    lambda_l1=0.1,
    lambda_l2=0.2,
)

# Train the model on the training data
lgb_regressor.fit(X_train_scaled, y_train)

# Predict on the test set
y_test_pred = lgb_regressor.predict(X_test_scaled)

# Calculate R-squared (R2) score for the test set
test_r2 = r2_score(y_test, y_test_pred)

print(f"Test R-squared: {test_r2:.4f}")
```

Nếu chúng ta không sử dụng phương pháp Hyperparameter Tuning thì model của chúng ta sẽ có Test R-Squared = 0.7966. Thực sự thì model cũng tương đối là tốt. Nhưng mà để tìm được các thông số của hyper-parameter làm cho model có performance tốt nhất thì thực sự là một thử thách khó. Nếu chúng ta sử dụng phương pháp Fine Tune, trong trường hợp này là sử dụng thư viện Optuna, thì chúng ta sẽ luôn luôn tìm ra được best hyper-parameter cho model.

### Sau khi dùng phương pháp Hyperparameter Tuning

```python
import lightgbm as lgb
import optuna

def objective(trial):
    params = {
        'boosting_type': 'gbdt',
        'objective': 'regression',
        'metric': 'rmse',
        'num_leaves': trial.suggest_int('num_leaves', 10, 100),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.001, 0.1),
        'feature_fraction': trial.suggest_uniform('feature_fraction', 0.1, 1.0),
        'bagging_fraction': trial.suggest_uniform('bagging_fraction', 0.1, 1.0),
        'bagging_freq': trial.suggest_int('bagging_freq', 1, 10),
        'lambda_l1': trial.suggest_loguniform('lambda_l1', 1e-8, 10.0),
        'lambda_l2': trial.suggest_loguniform('lambda_l2', 1e-8, 10.0),
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf', 1, 50),
        'random_seed': 42
    }

    model = lgb.LGBMRegressor(**params)

    model.fit(X_train_scaled, y_train, eval_set=[(X_test_scaled, y_test)])

    predict = model.predict(X_test_scaled)

    # Calculate R-squared (R2) score
    r2 = r2_score(y_test, predict)

    # Return the negative R2 score because Optuna minimizes the objective function
    return -r2  # Minimize negative R2
```

```python
# Create an Optuna study and run the optimization
study = optuna.create_study(direction='minimize')  # Minimize negative R2
study.optimize(objective, n_trials=100)

# Print the best hyperparameters and the best R2 score
print("Best trial:")
best_trial = study.best_trial
print("  Value (R2 Score): {:.4f}".format(-best_trial.value))  # Convert back to positive R2
print("  Params: ")
for key, value in best_trial.params.items():
    print("    {}: {}".format(key, value))
```

Như chúng ta có thể thấy, nếu chúng ta sử dụng phương pháp Fine Tune thì model của chúng ta tăng Test R-Squared từ 0.7966 lên 0.9494. Chúng ta có thể thấy rõ là việc sử dụng Fine Tune có thể làm performance của chúng ta cải thiện rất đáng kể.

# Optimization

### Không optimization


```python
lgb_regressor_reg = lgb.LGBMRegressor(
    objective='regression',
    boosting_type='gbdt',
    num_leaves= 46,
    learning_rate= 0.086,
    feature_fraction= 0.8,
    bagging_fraction = 0.6,
    bagging_freq =  2,
    min_data_in_leaf= 20,
    lambda_l1 = 0.1,
    lambda_l2 = 0.2)

lgb_regressor_reg.fit(X_train, y_train)
y_pred = lgb_regressor_reg.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

Sau khi chạy xong thì chúng ta có: Mean Squared Error = 11883.647200104566

### Using Hyperparameter Optimization (with Optuna):

```python
def objective(trial):
    params = {
        'boosting_type': 'gbdt',
        'objective': 'regression',
        'metric': 'rmse',
        'num_leaves': trial.suggest_int('num_leaves', 10, 100),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.001, 0.1),
        'feature_fraction': trial.suggest_uniform('feature_fraction', 0.1, 1.0),
        'bagging_fraction': trial.suggest_uniform('bagging_fraction', 0.1, 1.0),
        'bagging_freq': trial.suggest_int('bagging_freq', 1, 10),
        'lambda_l1': trial.suggest_loguniform('lambda_l1', 1e-8, 10.0),   # Regularization
        'lambda_l2': trial.suggest_loguniform('lambda_l2', 1e-8, 10.0),    # Regularization
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf', 1, 50),
        'random_seed': 42
    }

    model = lgb.LGBMRegressor(**params)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)

    return mse

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=100)   # Optimization

best_params = study.best_params
best_model = lgb.LGBMRegressor(**best_params)
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)
best_mse = mean_squared_error(y_test, y_pred)
print("Best Mean Squared Error:", best_mse)
```

Sau khi chạy xong, chúng ta thấy rõ Best Mean Squared Error là 9817.17 sau khi Optimization, khi không Optimization Mean Squared Error = 11883.647. Giảm MSE rõ rệt, model được cải thiện

# Model sau khi áp dụng 3 phương pháp: Hyperparameter tuning, Regularization and Optimization.

```python
def objective(trial):
    params = {
        'boosting_type': 'gbdt',
        'objective': 'regression',
        'metric': 'rmse',
        'num_leaves': trial.suggest_int('num_leaves', 10, 100),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.001, 0.1),
        'feature_fraction': trial.suggest_uniform('feature_fraction', 0.1, 1.0),
        'bagging_fraction': trial.suggest_uniform('bagging_fraction', 0.1, 1.0),
        'bagging_freq': trial.suggest_int('bagging_freq', 1, 10),
        'lambda_l1': trial.suggest_loguniform('lambda_l1', 1e-8, 10.0),
        'lambda_l2': trial.suggest_loguniform('lambda_l2', 1e-8, 10.0),
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf', 1, 50),
        'random_seed': 42
    }

    model = lgb.LGBMRegressor(**params)

    model.fit(X_train_scaled, y_train, eval_set=[(X_test_scaled, y_test)])

    predict = model.predict(X_test_scaled)

    # Calculate R-squared (R2) score
    r2 = r2_score(y_test, predict)

    # Return the negative R2 score because Optuna minimizes the objective function
    return -r2  # Minimize negative R2
```

```python
# Create an Optuna study and run the optimization
study = optuna.create_study(direction='minimize')  # Minimize negative R2
study.optimize(objective, n_trials=100)     # optimization

# Print the best hyperparameters and the best R2 score
print("Best trial:")
best_trial = study.best_trial
print("  Value (R2 Score): {:.4f}".format(-best_trial.value))  # Convert back to positive R2
print("  Params: ")
for key, value in best_trial.params.items():
    print("    {}: {}".format(key, value))
```

```python
lightgbm_model = lgb.LGBMRegressor(**best_trial.params)

# Train the model with early stopping
eval_set = [(X_train_scaled, y_train), (X_test_scaled, y_test)]
lightgbm_model.fit(
    X_train_scaled, y_train,
    eval_metric="rmse",
    eval_set=eval_set
)
```

```python
result = lightgbm_model.evals_result_  # Get evaluation results
train_error1 = result['training']['rmse']
val_error1 = result['valid_1']['rmse']

plt.figure(figsize=(7, 4))
plt.plot(train_error1, label='Train Error')
plt.plot(val_error1, label='Validation Error')
plt.xlabel('Number of Rounds')
plt.ylabel('RMSE')
plt.title('Train vs Validation error')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Review training results
y_pred_train_lightgbm = lightgbm_model.predict(X_train_scaled)

r2_lighttrain = r2_score(y_train, y_pred_train_lightgbm)
mse_lighttrain = mean_squared_error(y_train,y_pred_train_lightgbm)
mae_lighttrain = mean_absolute_error(y_train, y_pred_train_lightgbm)

print("R2 Score of train for LightGBM:", r2_lighttrain)
print("Mean Squared Error of train for LightGBM:", mse_lighttrain)
print("Mean Absolute Error of train for LightGBM:", mae_lighttrain)
```
